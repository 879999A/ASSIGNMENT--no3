Accept cost price(CP) and Selling price(SP) of an item. Calculate the ammount of PROFIT/LOSS made by the seller and print the
CP=500, SP=700	PROFIT=200
CP=500, SP=450	LOSS=50


STEPS - 1) TAKE COST PRICE AND SELLING PRICE FROM USER
2) IF COST PRICE IS GREATER THAN SELLING PRICE THAN IT WILL PRINT LOSS AMOUNT
3) IF COST PRICE IS LESS THAN SELLING PRICE THAN IT WILL PRINT PROFIT AMOUNT
4) ELSE IT WILL PRINT NO PROFIT NO LOSS


TEST_CASES - 
1) IF USER ENTERED ZERO VALUE
OUTPUT = please enter value grater than:  0
2) IF USER ENTERD STRING 
OUTPUT = please enter value only: 
3) IF USER ENTERED VALUE AS NEGATIVE 
OUTPUT = please enter value grater than:  0


```python
class Check_profit_loss():
    
    def __init__(self, cost_price, sale_price):
        self.cost_price = cost_price
        self.sale_price = sale_price
    
    def Profit_or_loss(self):
        if(cost_price > sale_price ):
            amount = cost_price - sale_price
            print("Total Loss Amount = {0} RS".format(amount))
        elif(sale_price > cost_price):
            amount = sale_price - cost_price
            print("Total Profit = {0} RS".format(amount))
        else:
            print("No Profit No Loss!!!")
    
        
if __name__ == '__main__':
    while True:
        try:
            cost_price = float(input("enter cost price"))
            assert cost_price >= 1
            sale_price = float(input(" enter sale_price"))
            assert sale_price >= 1
            break
        except ValueError:
            print(" please enter correct value")
        except AssertionError:
            print(" please enter value greater than zero")
    
    a = Check_profit_loss(cost_price,sale_price)
    a.Profit_or_loss()
    
```
